```python
if "setup_run" not in globals():
    raise SystemError("Please execute the setup cell before using this.")

from requests import get
import requests
from bs4 import BeautifulSoup
from time import sleep
from os.path import exists
import os  # Se necesita para usar os.stat
from json import load, dump
from google.colab import output
from ipywidgets import widgets
from jupyter_ui_poll import ui_events

!pip install jupyter_ui_poll --upgrade &> /dev/null

drive_path = '/content/drive/MyDrive/minecraft'
SERVERCONFIG = f'{drive_path}/server_list.txt'
# @title Create Server
# @markdown ####**Create your server.**
server_name = "" # @param {type:"string"}

# Cardboard
def get_mod_versions(mod):
    """
    Obtains the information of the available versions of a mod from Modrinth.
    """
    url = f"https://api.modrinth.com/v2/project/{mod}/version"
    response = requests.get(url)
    if response.ok:
        # Se retorna la lista completa de diccionarios de versi√≥n
        return response.json()
    else:
        ERROR(f"Error fetching {mod} versions from Modrinth.")

def get_mod_download_url(mod, version):
    """
    Searchs in the API for the download URL of a specific version of a mod.
    """
    url = f"https://api.modrinth.com/v2/project/{mod}/version"
    response = requests.get(url)
    if response.ok:
        data = response.json()
        for v in data:
            # version_number or name to search for the game version
            ver_num = v.get("version_number") or v.get("name")
            if ver_num == version:
                for file in v.get("files", []):
                    if file.get("filename", "").endswith(".jar"):
                        return file.get("url")
        ERROR(f"Specified {mod} version not found")
    else:
        ERROR(f"Error fetching {mod} download URL from Modrinth")

def install_fabric_and_cardboard(version):
    # Installs Fabric
    fabric_versions = SERVERSJAR("GetVersions", server_type="Cardboard")
    if not fabric_versions:
        ERROR("No Fabric versions found.")

    if version not in fabric_versions:
        ERROR(f"Fabric version {version} not found.")

    LOG(f"Installing Fabric {version}...")
    fabric_url = SERVERSJAR("GetDownloadUrl", server_type="fabric", version=version)

    os.makedirs(f'{drive_path}/{server_name}', exist_ok=True)
    fabric_jar_path = f'{drive_path}/{server_name}/fabric-server-{version}.jar'

    if not exists(fabric_jar_path):
        LOG(f"Downloading Fabric {version}...")
        os.system(f'wget -O "{fabric_jar_path}" "{fabric_url}"')
        os.rename(fabric_jar_path, f'{drive_path}/{server_name}/fabric-installer.jar')

    LOG(f"Fabric {version} installed successfully.")

    def install_required_mods(mod_id, mod_name, fabric_version):
        LOG(f"Installing {mod_name}")
        versions = get_mod_versions(mod_id)
        if not versions:
            ERROR(f"No versions found for {mod_name}.")

        # Searchs for the latest mod versions that matches the "game_versions"
        compatible_version = next(
            (v for v in versions if fabric_version in v.get("game_versions", [])),
            None
        )

        if not compatible_version:
            ERROR(f"No compatible {mod_name} version found for Fabric {fabric_version}.")

        mod_version_number = compatible_version.get("version_number") or compatible_version.get("name")
        LOG(f"Compatible {mod_name} version found: {mod_version_number}")
        mod_url = get_mod_download_url(mod_id, mod_version_number)
        mod_jar_path = f'{drive_path}/{server_name}/{mod_name}-{mod_version_number}.jar'

        if not exists(mod_jar_path):
            LOG(f"Downloading {mod_name} {mod_version_number}...")
            os.system(f'wget -O "{mod_jar_path}" "{mod_url}"')
            os.makedirs(f"{drive_path}/{server_name}/mods", exist_ok=True)
            os.rename(mod_jar_path, f"{drive_path}/{server_name}/mods/{mod_name}-{mod_version_number}.jar")

        LOG(f"{mod_name} {mod_version_number} installed successfully.")

    # Installs the two required mods to work with Cardboard and Carboard.
    install_required_mods("MLYQ9VGP", "Cardboard API", version)
    install_required_mods("P7dR8mSH", "Fabric API", version)
    install_required_mods("SVKv1SZo", "iCommon API", version)

#---------------------------------------------------------------------------------- SNALL FUNCTION ------------------------------------------------------------------------------------#
def SERVERSJAR(command, server_type=None, version=None):
    # Get the download URL (jar) AND return the detailed versions for each software (all)
    if command == "GetVersions":
        if server_type is None:
            ERROR("No server type specified.")
        Server_Jars_All = {
            'paper': 'https://api.papermc.io/v2/projects/paper',
            'velocity': 'https://api.papermc.io/v2/projects/velocity',
            'purpur': 'https://api.purpurmc.org/v2/purpur',
            'mohist': 'https://api.mohistmc.com/project/mohist/versions',
            'banner': 'https://api.mohistmc.com/project/banner/versions',
            'folia': 'https://api.papermc.io/v2/projects/folia'
        }
        if server_type in ['vanilla', 'snapshot']:
            rJSON = get('https://launchermeta.mojang.com/mc/game/version_manifest.json').json()
            if server_type == 'vanilla':
                server_type = 'release'
            if version != 'vanilla - latest_version':
                server_version = [hit["id"] for hit in rJSON["versions"] if hit["type"] == server_type]
            else:
                return rJSON['latest']['release']

        elif server_type in ['paper','velocity','purpur','folia']:
            rJSON = get(Server_Jars_All[server_type]).json()
            server_version = [hit for hit in rJSON["versions"]]

        elif server_type in ['mohist', 'banner']:
              rJSON = get(Server_Jars_All[server_type]).json()
              server_version = [v["name"] for v in rJSON]

        elif server_type == 'fabric':
            rJSON = get('https://meta.fabricmc.net/v2/versions/game').json()
            server_version = [hit['version'] for hit in rJSON if hit['stable'] == True]

        elif server_type == "neoforge":
            rJSON = get("https://maven.neoforged.net/api/maven/versions/releases/net/neoforged/neoforge").json()
            server_version = [f"1.{hit[:4]} - {hit}" for hit in rJSON["versions"]]

        elif server_type == 'forge':
            from bs4 import BeautifulSoup
            rJSON = get('https://files.minecraftforge.net/net/minecraftforge/forge/index.html')
            soup = BeautifulSoup(rJSON.content, "html.parser")
            server_version = [tag.text for tag in soup.find_all('a') if '.' in tag.text and '\n' not in tag.text]

        elif server_type == "bedrock":
            import requests
            DOWNLOAD_LINKS_URL = "https://net-secondary.web.minecraft-services.net/api/v1.0/download/links"
            BACKUP_URL = "https://raw.githubusercontent.com/ghwns9652/Minecraft-Bedrock-Server-Updater/main/backup_download_link.txt"
            HEADERS = {
                "User-Agent": "Mozilla/5.0 (X11; CrOS x86_64 12871.102.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.141 Safari/537.36"
            }
            try:
                response = requests.get(DOWNLOAD_LINKS_URL, headers=HEADERS, timeout=5)
                response.raise_for_status()
                all_links = response.json()['result']['links']
                download_link = next(
                    (link['downloadUrl'] for link in all_links if link['downloadType'] == 'serverBedrockLinux'),
                    None
                )

                if not download_link:
                    raise Exception("No 'serverBedrockLinux' link found in API response")

            except Exception:
                LOG("Failed to fetch from official API, using backup...")
                try:
                    response = requests.get(BACKUP_URL, headers=HEADERS, timeout=5)
                    response.raise_for_status()
                    download_link = response.text.strip()
                except Exception:
                    LOG("Backup URL also failed.")
                    download_link = None

            if download_link:
                try:
                    server_version = download_link.split('bedrock-server-')[1].split(".zip")[0]
                except Exception:
                    server_version = "unknown"
                    LOG("Could not extract version from download link.")
            else:
                server_version = None



        elif server_type == "arclight":
            LOG('Before going deeper, please check out https://github.com/IzzelAliz/Arclight')
            rJSON = get('https://files.hypoglycemia.icu/v1/files/arclight/minecraft').json()['files']
            server_version  = [hit['name'] for hit in rJSON]

        elif server_type == "crucible":
            server_version = ["1.7.10"]

        elif server_type == "magma":
            server_version = ["1.12.2", "1.18.2", "1.19.3", "1.20.1"]

        elif server_type == "ketting":
            server_version = ["1.20"]

        elif server_type == "cardboard":
            rJSON = requests.get("https://api.modrinth.com/v2/project/MLYQ9VGP/version").json()
            server_version = []
            for v in rJSON:
                if v['game_versions'][0] not in server_version:
                    server_version.append(v['game_versions'][0])
        elif server_type == "custom":
          LOG("Selected a custom software. Skipping version selection.")
          server_version = "custom"
        else:
            ERROR('Wrong server type.')
        return server_version

    elif command == "GetDownloadUrl":
        if version is None:
            ERROR("No version specified.")
        if server_type in ['vanilla', 'snapshot']:
            rJSON = get('https://launchermeta.mojang.com/mc/game/version_manifest.json').json()
            if server_type == 'vanilla':
                server_type = 'release'
            for hit in rJSON["versions"]:
                if hit["type"] == server_type and hit['id'] == version:
                    return get(hit['url']).json()["downloads"]['server']['url']

        elif server_type in ['paper','velocity','folia']:
            build = get(f'https://api.papermc.io/v2/projects/{server_type}/versions/{version}').json()["builds"][-1]
            jar_name = get(f'https://api.papermc.io/v2/projects/{server_type}/versions/{version}/builds/{build}').json()["downloads"]["application"]["name"]
            return f'https://api.papermc.io/v2/projects/{server_type}/versions/{version}/builds/{build}/downloads/{jar_name}'

        elif server_type == 'purpur':
            build = get(f'https://api.purpurmc.org/v2/purpur/{version}').json()["builds"]["latest"]
            return f'https://api.purpurmc.org/v2/purpur/{version}/{build}/download'

        elif server_type in ['mohist', 'banner']:
            builds_resp = get(f'https://api.mohistmc.com/project/{server_type}/{version}/builds').json()
            if not builds_resp:
                raise ValueError(f"[ERROR] No builds found for {server_type} version {version}")

            last_build_id = builds_resp[-1]["id"]
            return f'https://api.mohistmc.com/project/{server_type}/{version}/builds/{last_build_id}/download'

        elif server_type == 'fabric':
            installerVersion = get('https://meta.fabricmc.net/v2/versions/installer').json()[0]["version"]
            fabricVersion = get(f'https://meta.fabricmc.net/v2/versions/loader/{version}').json()[0]["loader"]["version"]
            return "https://meta.fabricmc.net/v2/versions/loader/" + version + "/" + fabricVersion + "/" + installerVersion + "/server/jar"

        elif server_type == 'forge':
            from bs4 import BeautifulSoup
            rJSON = get(f'https://files.minecraftforge.net/net/minecraftforge/forge/index_{version}.html')
            soup = BeautifulSoup(rJSON.content, "html.parser")
            tag = soup.find('a', title="Installer")
            tag = str(tag)
            tag = tag[tag.find('"') + 1 :]
            link = tag[:tag.find('"')]
            link = link[link.find('=') + 1:]
            link = link[link.find('=') + 1:]
            return link

        elif server_type == "neoforge":
            version = version.split(" - ")[-1]
            return f"https://maven.neoforged.net/releases/net/neoforged/neoforge/{version}/neoforge-{version}-installer.jar"

        elif server_type == 'arclight':
            rJSON = get(f'https://files.hypoglycemia.icu/v1/files/arclight/minecraft/{version}/loaders').json()
            LOG("Select your Arclight server type:")
            aldrop = widgets.Dropdown(description="Arclight Type:", options=[""]+[hit['name'] for hit in rJSON['files']])
            display(aldrop)
            with ui_events() as poll:
                while aldrop.value == '':
                    poll(10)
                    sleep(0.1)
            aldrop.close()
            build = aldrop.value
            LOG(f'Choosen Arclight type: {build}')
            LOG("Select your version type:")
            vtdrop = widgets.Dropdown(description="Arclight Type:", options=["", "Latest-snapshot", "Latest-stable"])
            display(vtdrop)
            with ui_events() as poll:
                while vtdrop.value == '':
                    poll(10)
                    sleep(0.1)
            vtdrop.close()
            choice = vtdrop.value
            LOG(f'Choosen version type: {choice}')
            if 'Latest-snapshot' == choice:
                choice = 'latest-snapshot'
            else:
                choice = "latest-stable"
            return f'https://files.hypoglycemia.icu/v1/files/arclight/minecraft/{version}/loaders/{build}/{choice}'

        elif server_type == 'bedrock':
            import requests
            DOWNLOAD_LINKS_URL = "https://net-secondary.web.minecraft-services.net/api/v1.0/download/links"
            BACKUP_URL = "https://raw.githubusercontent.com/ghwns9652/Minecraft-Bedrock-Server-Updater/main/backup_download_link.txt"
            HEADERS = {
                "User-Agent": "Mozilla/5.0 (X11; CrOS x86_64 12871.102.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.141 Safari/537.36"
            }

            try:
                response = requests.get(DOWNLOAD_LINKS_URL, headers=HEADERS, timeout=5)
                response.raise_for_status()
                all_links = response.json()['result']['links']
                download_link = next(
                    (link['downloadUrl'] for link in all_links if link['downloadType'] == 'serverBedrockLinux'),
                    None
                )

                if not download_link:
                    raise Exception("No 'serverBedrockLinux' link found in the API response.")

            except Exception as e:
                LOG(f"Primary source failed: {e}. Attempting backup...")
                try:
                    response = requests.get(BACKUP_URL, headers=HEADERS, timeout=5)
                    response.raise_for_status()
                    download_link = response.text.strip()
                except Exception as e_backup:
                    LOG(f"Backup source also failed: {e_backup}")
                    download_link = None

            if download_link:
                import os
                download_link_file = os.path.join(drive_path, server_name, 'updater', 'download_link.txt')
                os.makedirs(os.path.dirname(download_link_file), exist_ok=True)
                import os
                if not os.path.isfile(download_link_file):
                    with open(download_link_file, 'w') as f:
                        f.write(download_link)

            return download_link



        elif server_type == "magma":
            if version == "1.12.2":
                return "https://github.com/magmamaintained/Magma-1.12.2/releases/download/b4c01d2/Magma-1.12.2-b4c01d2-server.jar"
            elif version == "1.18.2":
                return "https://github.com/magmamaintained/Magma-1.18.2/releases/download/9938562/magma-1.18.2-40.2.21-9938562-server.jar"
            elif version == "1.19.3":
                return "https://github.com/magmamaintained/Magma-1.19.3/releases/download/0efc583/magma-1.19.3-44.1.23-0efc583-server.jar"
            elif version == "1.20.1":
                return "https://github.com/magmamaintained/Magma-1.20.1/releases/download/b4bdd80/magma-1.20.1-47.2.20-b4bdd80-server.jar"
            else:
                ERROR("The selected version is not on the list of available versions.")

        elif server_type == "crucible":
            return "https://github.com/CrucibleMC/Crucible/releases/download/staging-6b33785/Crucible-1.7.10-staging-6b337850-server.jar"

        elif server_type == "ketting":
            return "https://github.com/kettingpowered/kettinglauncher/releases/download/v1.5.1/kettinglauncher-1.5.1-sources.jar"

        elif server_type == "cardboard":
            # We use the external function to obtain Fabric and Cardboard
            return install_fabric_and_cardboard(version)
        elif server_type == "custom":
          return "server.jar"
        else:
            ERROR('Wrong server type.')
    elif command == "GetServerTypes":
        return ['Vanilla','Snapshot','Paper','Purpur','Mohist', "Arclight",'Velocity', 'Banner', 'Fabric',"Folia", 'Forge',"Neoforge", 'Bedrock', 'Crucible', 'Magma', 'Ketting', 'Cardboard', 'Custom']
    else:
        ERROR("Not a valid command.")

#---------------------------------------------------------------------------------- MAIN CODE ------------------------------------------------------------------------------------#
LOG(f"\nColab Version: {colabversion}")
# Auditing whether folder exists.
if exists(f'{drive_path}/{server_name}'):
    ERROR('Server already exists. If you want to change server software go to the "Change Server Software" cell.')

# Create folder
MKDIR(f'{drive_path}/{server_name}')
LOG('Checking if folder created. Please wait.')
sleep(10)
LOG("Select your server type:")
stdrop = widgets.Dropdown(description="Server Type:", options=[""]+SERVERSJAR("GetServerTypes"))
display(stdrop)
with ui_events() as poll:
    while stdrop.value == '':
        poll(10)
        sleep(0.1)
stdrop.close()
LOG(f'Choosen server type: {stdrop.value}')
server_type = stdrop.value.lower()

if server_type != "bedrock" and server_type != "custom":
    LOG("Select your server version:")
    svdrop = widgets.Dropdown(description="Server version: ", style={'description_width': 'initial'}, options=['']+SERVERSJAR("GetVersions", server_type=server_type))
    display(svdrop)
    with ui_events() as poll:
        while svdrop.value == '':
            poll(10)
            sleep(0.1)
    svdrop.close()
    version = svdrop.value
    LOG(f'\nChoosen server version: {version}')
elif server_type == "bedrock":
    version = SERVERSJAR("GetVersions", server_type=server_type)
    LOG("Using latest bedrock version")
else:
  from google.colab import files as fls
  uploaded = fls.upload()
  try:
    file_ = [fn for fn in uploaded.keys()][0] # Get the name of the uploaded file
    origin = os.path.join(drive_path, file_)
    destination = os.path.join(drive_path, server_name, 'server.jar')
    !mv "{origin}" "{destination}"
    version = "custom"
  except: ERROR("Lol, you haven't uploaded any file yet.")

LOG("Select a Tunnel provider:")
tunnelsrvs = widgets.Dropdown(description="Tunnel Service: ", style={'description_width': 'initial'}, options=['','Ngrok', 'Argo', 'Zrok', 'PlayIt', 'LocaltoNet', 'Tailscale', "Minekube-Gate"])
display(tunnelsrvs)
with ui_events() as poll:
    while tunnelsrvs.value == '':
        poll(10)
        sleep(0.1)
tunnelsrvs.close()
LOG(f'Choosen Tunnel service: {tunnelsrvs.value}')
tunnel_service = tunnelsrvs.value.lower()

# Checking if SERVERCONFIG has valid args
if not exists(SERVERCONFIG) or os.stat(SERVERCONFIG).st_size == 0:
    initial_config = {"server_list": [], "server_in_use": ""}
    with open(SERVERCONFIG, 'w') as f:
        dump(initial_config, f)
with open(SERVERCONFIG, 'r') as f:
    serverconfig = load(f)

serverconfig['server_list'] += [server_name]
serverconfig['server_in_use'] = server_name

# Tunnels info and token registration
tunnels_dashboard = {"ngrok":"https://dashboard.ngrok.com/get-started/your-authtoken", "zrok":"https://api.zrok.io", "localtonet":"https://localtonet.com/usertoken", "tailscale":"https://login.tailscale.com/admin/settings/keys . Make sure to select reusable", "minekube-gate":"https://app.minekube.com/orgs"}

if f"{tunnel_service}_proxy" not in serverconfig:
    serverconfig[f"{tunnel_service}_proxy"] = {"authtoken": ""}

if not serverconfig[f"{tunnel_service}_proxy"].get("authtoken") and not tunnel_service == 'playit':
  LOG("Tunnel Settings:")
  LOG(f'Get your authtoken from {tunnels_dashboard.get(tunnel_service)}')
  WARN("Remember to paste the whole authtoken at once!")

  adrop = widgets.Text(
      value='',
      placeholder='abc1234567890cba',
      description='Authtoken: ',
      disabled=False
  )
  display(adrop)
  with ui_events() as poll:
      while adrop.value == '' and len(adrop.value) < 2:
          poll(10)
          sleep(0.1)
  token = adrop.value
  adrop.close()
  # Stylish token output :D
  print_token = ''
  print_token = print_token + token[0:5]
  for i in range(0, len(token)-5): print_token = print_token + "*"
  LOG(f"Token saved: {print_token}")

  if token == "":
      ERROR("\nNo token provided")
  serverconfig[f"{tunnel_service}_proxy"]['authtoken'] = token

  # Tunnels extra conditionals
  if tunnel_service == 'ngrok':
      rdrop = widgets.Dropdown(description="Available regions: ", style={'description_width': 'initial'}, options=['','ap - Asia/Pacific (Singapore)', 'au - Australia (Sydney)', 'eu - Europe (Frankfurt - Germany)', 'in - India (Mumbai)', 'jp - Japan (Tokyo)', 'sa - America (S√£o Paulo - Brazil)', 'us - United States (Ohio)'])
      display(rdrop)
      with ui_events() as poll:
          while rdrop.value == '':
              poll(10)
              sleep(0.1)
      rdrop.close()
      region = rdrop.value.split(" ")[0]
      serverconfig['ngrok_proxy']['region'] = region
      LOG(f'Choosen region: {region}')
      rdrop.close()

  elif tunnel_service == 'tailscale':
      if "machine_info" not in serverconfig["tailscale_proxy"]:
          serverconfig["tailscale_proxy"]["machine_info"] = ""
else:
  serverconfig["playit_proxy"] = {"secretkey": ""}

dump(serverconfig, open(SERVERCONFIG, 'w'))

# Set up colabconfig
colabconfig = {"server_type": server_type, "server_version": version.split("-")[0], "tunnel_service": tunnel_service}
dump(colabconfig, open(COLABCONFIG(server_name),'w'))

# Download jar file
if server_type == "bedrock":
    DOWNLOAD_FILE(url=SERVERSJAR("GetDownloadUrl", server_type, version), path=f"{drive_path}/{server_name}", file_name='bedrock-server.zip', force=True, headers={"User-Agent":"Mozilla/4.0 (compatible; MSIE 6.0; Windows NT 5.1; BEDROCK-UPDATER)"})
    version = version.split("-")[0]
    LOG("Unzipping bedrock server zip.")
    ! unzip {drive_path}/{server_name}/bedrock-server.zip -d {drive_path}/{server_name} >/dev/null && $log "Successfully unzipped" || echo "unzip error."
    LOG("Unzipped successfully")
    sleep(10)
    LOG('\nCompleted!')
else:
    if server_type == 'forge':
        jarname = 'forge-installer.jar'
    elif server_type == 'neoforge':
        jarname = "neoforge-installer.jar"
    elif server_type == 'crucible':
        jarname = "Crucible-1.7.10-5.4.jar"
    elif server_type == 'ketting':
        jarname = "kettinglauncher-1.5.1-sources.jar"
    elif server_type == 'cardboard':
        jarname = "fabric-installer.jar"
    elif server_type == 'magma':
        jarname = "magma-installer.jar"
    elif server_type == 'custom':
        jarname = 'server.jar'
    else:
        jarname = JAR_LIST_RUN(version)[server_type]
    if server_type != "custom":
      DOWNLOAD_FILE(url=SERVERSJAR("GetDownloadUrl", server_type, version), path=f"{drive_path}/{server_name}", file_name=jarname)
      sleep(10)
    LOG('\nCompleted!')
```

```python
if "setup_run" not in globals(): raise SystemError("Please execute the setup cell before using this.")
from jproperties import Properties
from os.path import exists
from os import makedirs
import shutil
from datetime import datetime

# @markdown ####**Server Properties**
MaxPlayers = 0  # @param {type:"slider", min:0, max:100, step:1}
GameMode = "survival"  # @param ["survival", "creative", "adventure", "spectator"]
Difficulty = "normal"  # @param ["peaceful", "easy", "normal", "hard"]
Cracked = False  # @param {type:"boolean"}
PVP = False  # @param {type:"boolean"}
CommandBlock = False  # @param {type:"boolean"}
AllowFlight = False  # @param {type:"boolean"}
Animals = False  # @param {type:"boolean"}
Monsters = False  # @param {type:"boolean"}
Villagers = False  # @param {type:"boolean"}
Nether = False  # @param {type:"boolean"}
ForceGameMode = False  # @param {type:"boolean"}
SpawnProtection = 0  # @param {type:"slider", min:0, max:100, step:1}
LevelSeed = ""  # @param {type:"string"}

#---------------------------------------------------------------------------------- MAIN CODE ------------------------------------------------------------------------------------#

server_name = SERVER_IN_USE(server_name = '')
server_path = f"{drive_path}/{server_name}"
backup_world_dir = f"{drive_path}/backup/world"

if not exists(f"{server_path}/server.properties"):
    ERROR('Please run your Minecraft server before editing properties')
else:
    LOG('Updating server.properties...')
    server_properties = Properties()
    with open(f"{server_path}/server.properties", "rb") as f:
        server_properties.load(f, "utf-8")

    server_properties["max-players"] = str(MaxPlayers)
    server_properties["gamemode"] = GameMode
    server_properties["difficulty"] = Difficulty

    clean_seed = LevelSeed.strip()

    if clean_seed != "":
        makedirs(backup_world_dir, exist_ok=True)

        world_paths = {
            "world": f"{server_path}/world",
            "world_nether": f"{server_path}/world_nether",
            "world_the_end": f"{server_path}/world_the_end"
        }
        existing_worlds = [name for name, path in world_paths.items() if exists(path)]

        if existing_worlds:
            timestamp = datetime.now().strftime("%Y%m%dT%H%M%S")
            seed_for_name = clean_seed.replace(" ", "_")
            backup_name = f"{server_name}_worlds_level-seed-{seed_for_name}_{timestamp}"
            backup_path = f"{backup_world_dir}/{backup_name}"
            makedirs(backup_path, exist_ok=True)

            for world in existing_worlds:
                LOG(f"Backing up: {world}")
                shutil.copytree(f"{server_path}/{world}", f"{backup_path}/{world}")
                LOG(f"Completed: {world}")

            LOG(f"Backup completed at {backup_path}")

            # Delete original worlds
            for world in existing_worlds:
                shutil.rmtree(f"{server_path}/{world}")

    server_properties["level-seed"] = clean_seed

    settings_dict = {
        'pvp': PVP,
        'enable-command-block': CommandBlock,
        'allow-flight': AllowFlight,
        'spawn-animals': Animals,
        'spawn-monsters': Monsters,
        'spawn-npcs': Villagers,
        'allow-nether': Nether,
        'force-gamemode': ForceGameMode,
        'spawn-protection': SpawnProtection,
        "online-mode": not Cracked
    }
    for key, value in settings_dict.items():
        server_properties[key] = str(value).lower()

    with open(f"{server_path}/server.properties", "wb") as f:
        server_properties.store(f, encoding="utf-8")
    LOG('Completed')
```
